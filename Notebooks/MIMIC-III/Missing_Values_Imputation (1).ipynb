```python
df=pd.read_csv("delirium.csv")
df
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
cols=list(df)
missing_vals_tracker=pd.DataFrame()
missing_vals_tracker['Feature']=cols
missing_vals=[]
for col in cols:
  miss_percent=(df[col].isnull().sum()/df.shape[0])*100
  missing_vals.append(miss_percent)
missing_vals_tracker['Missing Values %']=missing_vals
```

```python
missing_vals_tracker
```

```python
missing_vals_tracker.to_csv("Delirium_Missing_Values_Percentages.csv")
```

```python
df_miss_imputed_tr,df_miss_imputed_ts=dp.pipeline(df,'Delirium',prepare_f=True,split_f=True,impute_f=True,downsample_f=False,threshold=0.60,factor=1.5)
```

```python
df_miss_imputed_tr.to_csv("delirium_train_missingImputed.csv")
df_miss_imputed_ts.to_csv("delirium_test_missingImputed.csv")
```
